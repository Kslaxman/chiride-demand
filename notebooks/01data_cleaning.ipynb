##### Rename Columns

```python
df = df_raw.copy()

df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('percent', 'pct')
)

df.columns
```

##### Fix Data Types in Currency Columns

```python
currency_cols  = ['fare', 'tip', 'additional_charges', 'trip_total']

for col in currency_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )

    df[col] = df[col].replace('nan', np.nan)
    df[col] = df[col].replace('', np.nan)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

##### Fix Percentage Column

```python
pct_cols = ['pct_time_chicago', 'pct_distance_chicago']

for col in pct_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('%', '', regex=False)
        .str.strip()
    )

    df[col] = df[col].replace('nan', np.nan)
    df[col] = df[col].replace('', np.nan)
    df[col] = pd.to_numeric(df[col], errors='coerce') / 100
```

##### Fix DateTime Columns

```python
datetime_cols = ['trip_start_timestamp', 'trip_end_timestamp']

"""
%m = month (12)
%d = day (31)
%Y = 4-digit year (2024)
%I = 12-hour clock hour (11)
%M = minute (45)
%S = second (00)
%p = AM/PM

"""

for col in datetime_cols:
    df[col] = pd.to_datetime(df[col], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
    print(df[col].head())
```

##### Verify Date Range

```python
start_col = 'trip_start_timestamp'
print(f"\nEarliest trip start: {df[start_col].min()}")
print(f"Latest trip start:   {df[start_col].max()}")
print(f"Date range span:     {(df[start_col].max() - df[start_col].min()).days} days")

# Distribution by date
print(f"\nTrips per day (first 5 and last 5):")
daily_counts = df[start_col].dt.date.value_counts().sort_index()
print("\nFirst 5 days:")
print(daily_counts.head().to_string())
print("\nLast 5 days:")
print(daily_counts.tail().to_string())
print(f"\nTotal unique dates: {daily_counts.shape[0]}")
# print(daily_counts)
```

##### Drop Rows Missing Critical Data

```python
critical_cols = ['trip_start_timestamp', 'pickup_centroid_latitude', 'pickup_centroid_longitude']
critical_nulls = df[critical_cols].isnull().any(axis=1).sum()
print(f"Rows missing critical data: {critical_nulls:,}")

df = df.dropna(subset=critical_cols).reset_index(drop=True)
```

##### Drop Duplicate Trips

```python
duplicate_count = df['trip_id'].duplicated().sum()

if duplicate_count > 0:
    df = df.drop_duplicates(subset='trip_id', keep='first').reset_index(drop=True)
else:
    print("No duplicate trip_id values found.")
```

##### Remove Outliers

```python
# Filter Duration
df['trip_seconds'] = pd.to_numeric(df['trip_seconds'], errors='coerce')
mask_duration = (df['trip_seconds'] > 0) & (df['trip_seconds'] <= 10800)
removed_duration = (~mask_duration).sum()
print(f"\nRemoving {removed_duration:,} trips with invalid durations.")

# Filter Distance
mask_distance = (df['trip_miles'] > 0) & (df['trip_miles'] <= 200)
removed_distance = (~mask_distance).sum()
print(f"Removing {removed_distance:,} trips with invalid distances.")

# Filter Fare
mask_fare = (df['fare'] >= 0) & (df['fare'].notna())
removed_fare = (~mask_fare).sum()
print(f"Removing {removed_fare:,} trips with invalid fare amounts.")

# Pickup location bounds within chicago metro area
mask_lat = (df['pickup_centroid_latitude'] >= 41.6) & (df['pickup_centroid_latitude'] <= 42.1)
mask_lon = (df['pickup_centroid_longitude'] >= -88.0) & (df['pickup_centroid_longitude'] <= -87.4)
mask_location = mask_lat & mask_lon
removed_location = (~mask_location).sum()
print(f"Removing {removed_location:,} trips with pickup locations outside Chicago metro area.")

mask = mask_duration & mask_distance & mask_fare & mask_location
df = df[mask].reset_index(drop=True)
```

##### Drop Unnecessary Columns

```python
cols_to_drop = ['trip_id', 'pickup_centroid_location', 'dropoff_centroid_location']

existing_drops = [col for col in cols_to_drop if col in df.columns]
missing_drops = [col for col in cols_to_drop if col not in df.columns]

df = df.drop(columns=existing_drops)
```
